RMedian : Phase 1 & 2

```python
import math
import random
import statistics
```

Testfall :

```python
# User input
n = 2**14 + 1
case = 1

# Cases
if case == 1: #
    k = int(n ** (2 / 3))
    d = int(n ** (1 / 12))
elif case == 2:
    k = int(n / (math.log(n) / math.log(2)))
    d = int(math.log(n) / math.log(2))
    
# Automatic
X = [i for i in range(n)]
cnt = [0 for _ in range(n)]

# Print
print(n, k, d)
```

Algorithmus : Phase 1 & 2

```python
def phase1(X, k, d):

    n = len(X)
    med = statistics.median(X)

    random.shuffle(X)
    S = X[:k]
    XS = X[k:]

    S.sort()

    if 2*(k*math.log2(n))**0.5 < k/2:
        lst = [2*(k*math.log2(n))**0.5]
        if 3*(k*math.log2(n))**0.5 < k/2:
            lst.append(3*(k*math.log2(n))**0.5)
            while d*lst[len(lst) - 1] < k/2:
                lst.append(d*lst[len(lst) - 1])
        lst.append(k/2)
    else:
        lst = [k/2]

    L = [[] for _ in range(len(lst) - 1)]
    R = [[] for _ in range(len(lst) - 1)]
    C = []

    for s in S[math.floor(k / 2 - lst[0]): math.ceil(k / 2 + lst[0])]:
        C.append(s)

    for i in range(1, len(lst)):
        for s in S[math.floor(k / 2 - lst[i]): math.floor(k / 2 - lst[i - 1])]:
            L[i - 1].append(s)
        for s in S[math.ceil(k / 2 + lst[i - 1]): math.ceil(k / 2 + lst[i])]:
            R[i - 1].append(s)

    return S, XS, L, C, R, lst

# ========================================
def phase2(S, XS, L, C, R, cnt):
    mark = [False for _ in range(2**20)]
    b = len(L)

    random.shuffle(XS)
    for x_i in XS:
        med = 0
        for j in reversed(range(0, b - 1)):
            
            current = 2**50
            random.shuffle(L[j])
            for l in L[j]:
                if cnt[l] < current:
                    x_A = l

            if mark[x_A] == True:
                c = 1
            
            else:
                c = 2

            cnt[x_i] += 1
            cnt[x_A] += 1
            
            if x_i < x_A:
                if j + c < b:
                    mark[x_i] = True
                    L[j + c].append(x_i)
                    med = -1
                else:
                    med = -2
                break

            current2 = 2**50
            random.shuffle(R[j])
            for r in R[j]:
                if cnt[r] < current2:
                    x_B = r

            if mark[x_B] == True:
                c = 1
            else:
                c = 2

            cnt[x_i] += 1
            cnt[x_B] += 1
            
            if x_i > x_B:
                if j + c < b:
                    mark[x_i] = True
                    R[j + c].append(x_i)
                    med = 1
                else:
                    med = 2
                break
        if med == 0:
            C.append(x_i)
        elif med == -2:
            L[len(L) - 1].append(x_i)
        elif med == 2:
            R[len(R) - 1].append(x_i)
            
            
    sumL, sumR = 0, 0
    for l in L:
        sumL += len(l)
    for r in R:
        sumR += len(r)

    s = sumL - sumR
            
        #   Expand
    if s < 0:
        rs = []
        for r in R:
            rs += r
        random.shuffle(rs)
        for i in range(-s):
            C.append(rs[i])
            for r in R:
                if rs[i] in r:
                    r.remove(rs[i])
    else:
        ls = []
        for l in L:
            ls += l
        random.shuffle(ls)
        for i in range(s):
            C.append(ls[i])
            for l in L:
                if ls[i] in l:
                    l.remove(ls[i])
    
    return S, XS, L, C, R, cnt

# ========================================
# Testfall
S, XS, L, C, R, lst = phase1(X, k, d)
S, XS, L, C, R, cnt = phase2(S, XS, L, C, R, cnt)

S, XS, L, C, R, lst = phase1(C, k, d)
S, XS, L, C, R, cnt = phase2(S, XS, L, C, R, cnt)

S, XS, L, C, R, lst = phase1(C, k, d)
S, XS, L, C, R, cnt = phase2(S, XS, L, C, R, cnt)
S, XS, L, C, R, lst = phase1(C, k, d)
S, XS, L, C, R, cnt = phase2(S, XS, L, C, R, cnt)
S, XS, L, C, R, lst = phase1(C, k, d)
S, XS, L, C, R, cnt = phase2(S, XS, L, C, R, cnt)
print('done')
```

Resultat :

```python
def test(X, k, d, S, XS, L, C, R, lst, cnt):
    X.sort()
    med = statistics.median(X)
    n, s, xs, l, c, r, sumL, sumR = len(X), len(S), len(XS), len(L), len(C), len(R), 0, 0
    
    for i in range(len(lst)):
        lst[i] = round(lst[i], 2)
    
    for i in range(len(L)):
        sumL += len(L[i])
        sumR += len(R[i])
    C.sort()
    
    print('')
    print('Testfall n / k / d:', n, '/', k, '/', d)
    print('=======================================')
    print('|S| / |XS|  :', s, '/', xs)
    print('=======================================')
    print('n_i :', lst)
    for i in reversed(range(len(lst) - 1)):
        print('L' +str(i+1), ' :', '[', int(k/2 - lst[i + 1]), ',', int(k/2 - lst[i]), ']')
    print('C   :', '[', int(k/2 - lst[0]), ',', int(k/2 + lst[0]), ']')
    for i in range(len(lst) - 1):
        print('R' +str(i+1), ' :', '[', int(k/2 + lst[i]), ',', int(k/2 + lst[i + 1]), ']')
    print('---------------------------------------')
    print('||L|| :', sumL)
    for i in reversed(range(len(L))):
        print('|L'+str(i+1)+'|  :', len(L[i]))
    print('|C|   :', c)
    for i in range(len(L)):
        print('|R'+str(i+1)+'|  :', len(R[i]))
    print('||R|| :', sumR)
    print('---------------------------------------')
    print('|L + C + R| :', sumL + c + sumR)
    print('=======================================')
    print('Median  :', med)
    print('Med in C:', med in C)
    print('Med is Median from C :', statistics.median(C) == med)
    print(med)
    print(statistics.median(C))
    print('=======================================')
    print('max(cnt)     :', max(cnt))
    return

# Testfall
test(X, k, d, S, XS, L, C, R, lst, cnt)
```
